```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(convert) #this line transforms the 'genres' column by converting the string representation of genres
# in each row into a list of 'name' values using the 'convert' function.
```

```python
movies.head()
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x]) ## [i.replace(" ","") for i in x]: This is a list comprehension inside the lambda function.
#                                  It iterates over each string in the list 'x' (genres),
#                                  and for each string 'i', it removes spaces using the 'replace' method.
#                                  The result is a new list of strings with spaces removed.
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
movies.head()
```

```python
new_df=movies[['movie_id','title','tags']]
new_df.head()
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x)) # " ".join(x): This is the expression inside the lambda function.
#               It joins the elements of the list 'x' (tags) into a single string using the 'join' method.
#               The elements are joined with a space (' ') as the separator.
```

```python
new_df.head()
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
import nltk 
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
#The PorterStemmer is used for stemming, a process in natural language processing that reduces words to their base or root form.
```

```python
def stem(text):
    l=[]
    for i in text.split():
        l.append(ps.stem(i)) # Apply stemming to each word using the PorterStemmer ('ps')
    ## Join the list of stemmed words into a string using a space as the separator
    # This results in a string where each word is the stemmed version of the original word
    return " ".join(l)
        
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english') 
# Overall, this code sets up a CountVectorizer with a maximum vocabulary size of 5000 and
# specifies that common English stop words should be ignored during the tokenization process.
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray() 
# Overall, this line of code transforms the text data in the 'tags' column into a matrix of token counts
# using the CountVectorizer, and the result is stored in the 'vector' variable.
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6] 
# sorting the elements of the enumerated list enumerate(similarity[0]) based on the second element of each tuple (index 1, which is the similarity score), in descending order. It then selects the top 5 elements (excluding the first one), which effectively gives you the indices of the most similar items to the first item.
```

```python
def recommend(movie):
    # Find the index of the given movie in the DataFrame 'new_df'
    movie_index = new_df[new_df['title'] == movie].index[0]

    # Extract the similarity scores for the given movie
    distances = similarity[movie_index]

    # Create a list of tuples where each tuple contains an index and its corresponding similarity score
    movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])

    # Print the titles of the top 5 recommended movies (excluding the input movie itself)
    for i in movie_list[1:6]:
        recommended_movie_title = new_df.iloc[i[0]].title
        print(recommended_movie_title)

# Example usage:
# recommend("The Matrix")

        
```

```python
recommend('Avatar')
```
